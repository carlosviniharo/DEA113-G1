```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/netflix_1.csv'
data = pd.read_csv(url)
data.head()
```

```python
# Selecting only the data that the type is Movie.
data = data[data["type"] == 'Movie']
```

```python
print(data['type'].unique())
```

```python
data.nunique()
```

```python
data.isnull().sum()
```

```python
data['duration'] = data['duration'].fillna('min 0')
```

```python
data.isnull().sum()
```

```python
# Extract numeric values from the 'duration' column
data['duration'] = data['duration'].str.extract(r'(\d+)').astype(int)
```

```python
data.head()
```

```python
data['duration'][data['duration'] == 0]
```

```python
# Replacing the average value for values with value zero
data['duration'] = data['duration'].replace(0, data['duration'][data['duration'] != 0].mean())
```
